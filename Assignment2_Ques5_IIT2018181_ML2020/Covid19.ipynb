```python
x_train = train_set["Serial number"].copy()
y_train = train_set[" New_deaths"].copy()
x_test = test_set["Serial number"].copy()
y_test = test_set[" New_deaths"].copy()
```

```python
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
```

```python
#Linear Curve Fitting
```

```python
sumX = sum(x_train)
```

```python
sumY = sum(y_train)
```

```python
squareX = []
for i in range(len(x_train)):
    squareX.append(x_train[i]**2)
sumSquareX = sum(squareX)
```

```python
XY = []
for i in range(len(x_train)):
    XY.append(x_train[i]*y_train[i])
sumXY = sum(XY)
```

```python
w0 = (sumX*sumXY-sumY*sumSquareX)/(sumX**2 - sumSquareX*22)
```

```python
w1 = (sumX*sumY - sumXY*22)/(sumX**2 - sumSquareX*22 )
```

```python
w0
```

```python
w1
```

```python
juneTen = 12157 #prediction of June 10th
pred1 = w0 + w1*juneTen
pred1
```

```python
aprilTwenty = 12106 #prediction of April 20th
pred2 = w0 + w1*aprilTwenty
pred2
```

```python
meanList = []
for i in range(len(x_train)):
    meanList.append(((w0+w1*x_train[i])-y_train[i])**2)
accuracy = mean(meanList)
    
```

```python
accuracy# Mean square error for linear
```

```python
cubeX = []
for i in range(len(x_train)):
    cubeX.append(x_train[i]**3)
sumCubeX = sum(cubeX)
```

```python
quarticX = []
for i in range(len(x_train)):
    quarticX.append(x_train[i]**4)
sumQuarticX = sum(quarticX)
```

```python
xSquareY = []
for i in range(len(x_train)):
    xSquareY.append(((x_train[i])**2)*y_train[i])
sumXSquareY = sum(xSquareY)
```

```python
#Quadratic Curve Fitting
```

```python
w0,w1,w2 = util.equationInThreeVariables((22,sumX,sumSquareX,sumY),(sumX,sumSquareX,sumCubeX,sumXY),(sumSquareX,sumCubeX,sumQuarticX,sumXSquareY))
```

```python
w0
```

```python
w1
```

```python
w2
```

```python
predictionList = []
for i in range(len(x_train)):
    actualY = y_train[i]
    X = x_train[i]
    predictedY = w0 + w1*X + w2*(X**2)
    predictionList.append([actualY,predictedY])
meanList = []
for var in predictionList:
    meanList.append((var[1]-var[0])*(var[1]-var[0]))
accuracy = mean(meanList)

```

```python
accuracy# Mean square error for Quadratic
```
